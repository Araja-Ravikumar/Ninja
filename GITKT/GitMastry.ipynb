# What are the Different options we have in this situation

### The Problem
1. Then, you tried `git remote add origin <url>`, but Git says `origin` already exists.

### Solution

#### 1. First, check your existing remotes:
```bash
git remote -v
```
This will show if `origin` is already configured (likely with a different URL).

------------------

#### 2. You have 3 options:

**Option A: Update the existing remote URL**
```bash
git remote set-url origin https://github.com/Araja-Ravikumar/Ninja.git
```

**Option B: Remove the old remote first, then add the new one**
```bash
git remote remove origin
git remote add origin https://github.com/Araja-Ravikumar/Ninja.git
```

**Option C: Add a different remote name (if you want to keep both)**
```bash
git remote add new-origin https://github.com/Araja-Ravikumar/Ninja.git
```

---------------------

# checking the setup

#### 3. Verify it worked:
```bash
git remote -v
```

---------------

### Common Next Steps:
```bash
git push -u origin main  # If pushing to main branch for the first time
# or
git push -u origin master  # If using master branch
```

This error occurs because you don't have **write permissions** for the `Araja-RaviKumar/Ninja.git` repository. Let's fix this:

### Possible Reasons & Solutions:

#### 1. **You Don't Have Push Access**
   - The repository belongs to `Araja-RaviKumar`, and GitHub is denying access to your account (`mohan-araja`).
   - **Solution:** Ask the owner (`Araja-RaviKumar`) to add you as a **collaborator** in:
     - **Repo Settings → Collaborators → Add people**

#### 2. **Using HTTPS with Wrong Credentials**
   - Git is likely using cached credentials from a different account.
   - **Fix:** Update your Git credentials:

   **On Windows:**
   ```bash
   git credential-manager reject https://github.com
   ```
   Then try pushing again—it will prompt for GitHub credentials.

   **On Mac/Linux:**
   ```bash
   git config --global credential.helper store
   git push -u origin master  # Will ask for username/password
   ```

#### 3. **Alternative: Use SSH Instead of HTTPS**
   - More reliable if you have SSH keys set up.
   - **Steps:**
     ```bash
     git remote set-url origin git@github.com:Araja-RaviKumar/Ninja.git
     git push -u origin master
     ```
     *(Requires [SSH key setup](https://docs.github.com/en/authentication/connecting-to-github-with-ssh))*

#### 4. **Fork the Repository (If You Can’t Get Access)**
   - If you can't get collaborator access, fork the repo:
     1. Go to https://github.com/Araja-RaviKumar/Ninja
     2. Click **Fork** (top-right)
     3. Clone your fork instead:
        ```bash
        git remote set-url origin https://github.com/mohan-araja/Ninja.git
        git push -u origin master
        ```

### Quick Test:
Run this to see which account Git is using:
```bash
curl -H "Authorization: token $(git config --global github.token)" https://api.github.com/user
```
*(If you get a 401, credentials are incorrect.)*
