

# <span style='color:#2980b9'>Pre-Learning Plan for Absolute Beginners</span>
### <span style='color:#7f8c8d'>From Zero to Programming Logic</span>

---

## <span style='color:#e74c3c'>1. What is a Program?</span>
<span style='color:#2c3e50'>**Core Concept:**</span>  
A program is like a cooking recipe - a set of **step-by-step instructions** for computers.

<span style='color:#27ae60'>**Real-World Example:**</span>  
```plaintext
Making Tea:
1. Boil water
2. Add tea leaves
3. Add sugar
4. Pour into cup
```

<span style='color:#f39c12'>**Key Insight:**</span>  
- Computers follow instructions **literally**  
- Wrong order (e.g., pouring before boiling) → Bad result  
- Missing steps → Program fails

---

## <span style='color:#e74c3c'>2. Variables: Storing Information</span>
<span style='color:#2c3e50'>**Core Concept:**</span>  
Variables are **labeled containers** that hold information.

<span style='color:#27ae60'>**Visual Example:**</span>  
```plaintext
[Name]    → Contains "Priya"
[Age]     → Contains 28
[Student] → Contains True
```

<span style='color:#f39c12'>**Hands-On Activity:**</span>  
Create your own variables:  
- `Favorite_Food = "Pizza"`  
- `Siblings = 2`  
- `Has_Pet = True`

---

## <span style='color:#e74c3c'>3. Making Decisions (If/Else)</span>
<span style='color:#2c3e50'>**Core Concept:**</span>  
Programs choose actions based on conditions.

<span style='color:#27ae60'>**Everyday Scenario:**</span>  
```plaintext
WEATHER = "Rainy"

IF WEATHER == "Rainy":
    Take umbrella
ELSE:
    Wear sunglasses
```

<span style='color:#f39c12'>**Practice Exercise:**</span>  
Create rules for:  
- `IF score >= 90 THEN print("A grade") ELSE study_more()`  
- `IF tired THEN sleep ELSE keep_working()`

---

## <span style='color:#e74c3c'>4. Loops: Repeating Actions</span>
<span style='color:#2c3e50'>**Core Concept:**</span>  
Avoid repetition by creating loops.

<span style='color:#27ae60'>**Physical Demonstration:**</span>  
```plaintext
REPEAT 5 TIMES:
    Clap hands
    Jump once
```

<span style='color:#f39c12'>**Activity:**</span>  
Convert to loops:  
- "Wash dishes until all clean" → `while dishes_dirty: wash()`  
- "Count from 1 to 10" → `for number in 1 to 10: print(number)`

---

## <span style='color:#e74c3c'>5. Functions: Reusable Code Blocks</span>
<span style='color:#2c3e50'>**Core Concept:**</span>  
Functions are like kitchen appliances - define once, use often.

<span style='color:#27ae60'>**Analogy:**</span>  
```plaintext
FUNCTION Make_Smoothie(fruit):
    1. Add fruit to blender
    2. Add milk
    3. Blend for 30 seconds
    4. Pour into glass
```

<span style='color:#f39c12'>**Create Your Own:**</span>  
Define these functions:  
- `Morning_Routine()` with 3 steps  
- `Calculate_Area(length, width)`

---

## <span style='color:#e74c3c'>6. Lists: Grouping Items</span>
<span style='color:#2c3e50'>**Core Concept:**</span>  
Lists are collections like shopping baskets.

<span style='color:#27ae60'>**Practical Example:**</span>  
```plaintext
Groceries = ["Milk", "Eggs", "Bread", "Cheese"]
Even_Numbers = [2, 4, 6, 8, 10]
```

<span style='color:#f39c12'>**Exercise:**</span>  
Create lists for:  
- Your top 5 favorite songs  
- All family members' names  
- Weekday names
